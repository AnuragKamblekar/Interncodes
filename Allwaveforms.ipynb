```
%matplotlib inline

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter
from sklearn.preprocessing import StandardScaler
from keras import layers, models

class NissanLeafAnalyzerCNN:
    """Hybrid Analyzer using Savitzky-Golay + CNN regression."""

    def __init__(self, csv_name="denoisedsignal.csv"):
        self.input_path = "sensor_raw.csv"   # upload this file in Colab
        self.csv_name = csv_name
        self.columns = ['AccX','AccY','AccZ','GyroX','GyroY','GyroZ']

    def apply_savgol_filter(self, signal, window=11, polyorder=3):
        return savgol_filter(signal, window_length=window, polyorder=polyorder)

    def build_cnn(self, input_shape):
        # Regression CNN (Dense network for per-point regression)
        model = models.Sequential([
            layers.Input(shape=input_shape),
            layers.Dense(32, activation='relu'),
            layers.Dense(64, activation='relu'),
            layers.Dense(1, activation='linear')   # continuous output
        ])
        model.compile(optimizer='adam', loss='mse', metrics=['mae'])
        return model

    def run_analysis(self):
        if not os.path.exists(self.input_path):
            print(f"Error: {self.input_path} not found.")
            return

        # Load first 100 rows
        df_raw = pd.read_csv(self.input_path).head(100)
        time = np.arange(len(df_raw))

        for col in self.columns:
            if col not in df_raw.columns:
                continue

            # Raw and denoised signals
            raw_data = df_raw[col].values
            denoised = self.apply_savgol_filter(raw_data)

            # Scale raw input
            scaler = StandardScaler()
            X = scaler.fit_transform(raw_data.reshape(-1,1))
            X = X.reshape((len(X), 1))   # 100 samples, 1 feature

            # Target = denoised waveform (continuous regression)
            y = denoised.reshape((len(denoised), 1))   # 100 targets, 1 feature

            # Build regression CNN
            model = self.build_cnn((1,))
            model.fit(X, y, epochs=20, verbose=1)

            # Predictions (continuous waveform, length 100)
            predictions = model.predict(X).flatten()
            labels = predictions

            # Plot all outputs
            self._plot_all(col, time, raw_data, denoised, labels)

        plt.show()

    def _plot_all(self, col_name, time_axis, raw_signal, denoised, labels):
        fig, axs = plt.subplots(5, 1, figsize=(14, 20), sharex=True)
        fig.suptitle(f"CNN Hybrid Analysis Profile: {col_name}", fontsize=16, fontweight='bold')

        # 1. Raw dataset
        axs[0].plot(time_axis, raw_signal, color='gray')
        axs[0].set_title("Raw Dataset")

        # 2. Denoised dataset
        axs[1].plot(time_axis, denoised, color='orange')
        axs[1].set_title("Denoised Dataset")

        # 3. Savitzky-Golay Filter (applied on denoised)
        axs[2].plot(time_axis, self.apply_savgol_filter(denoised), color='purple')
        axs[2].set_title("Savitzky-Golay Filter Output")

        # 4. CNN Output (waveform regression)
        axs[3].plot(time_axis, labels, color='blue', linestyle='--')
        axs[3].set_title("CNN Reconstructed Waveform")

        # 5. Superimposed comparison
        axs[4].plot(time_axis, raw_signal, color='gray', alpha=0.3, label='Raw')
        axs[4].plot(time_axis, denoised, color='orange', label='Denoised')
        axs[4].plot(time_axis, self.apply_savgol_filter(denoised), color='purple', label='Savitzky-Golay')
        axs[4].plot(time_axis, labels, color='blue', linestyle='--', label='CNN Waveform')
        axs[4].legend()
        axs[4].set_title("Superimposed Comparison")

        plt.show()

if __name__ == "__main__":
    analyzer = NissanLeafAnalyzerCNN()
    analyzer.run_analysis()
```
